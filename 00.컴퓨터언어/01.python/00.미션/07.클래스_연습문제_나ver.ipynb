```python
# 은행잔고확인 통장 미션

class CAccount00 :
    def __int__(self) :
        self.amt = 0   
        
    def deposit() :
        self.deposit_amt = int(input('입금할 금액을 입력해주세요: '))
        if self.deposit_amt <= 0 :
            print('정확한 금액을 입력하세요.')
            continue
        else:
            self.amt += self.deposit_amt
        return self.amt
    
    def withdraw() :
        self.withdraw_amt = int(input('출금할 금액을 입력해주세요: '))
        if self.withdraw_amt > self.amt :
            print('잔액부족, 거래가 거절되었습니다')
            break
        if self.withdraw_amt <= 0:
            print('정확한 금액을 입력하세요.')
            continue
        else:
            self.amt -= self.withdraw_amt
        return self.amt
```

```python
# 교수님과 함께하는 클래스 선언 skeleton

class CAccount:
    def __init__(self,amount = 0):
        self.amount = amount
        self.process = True
        
    def bankTransaction(self):
        message = '''
---------------
1. 잔액확인
2. 입금
3. 출금
4. 종료
---------------
'''
        while type(message) == str:
            print(message)
            print()
            inum = int(input("메뉴를 선택해주세요: "))
            if inum == 4:
                print('선택 메뉴: 4. 종료')
                print('계좌 거래가 종료되었습니다.')
                break # return -1
            if inum == 1:
                print('선택 메뉴: 1. 잔액확인')
                self.disBalance()
                continue # return 1
            elif inum == 2:
                print('선택 메뉴: 2. 입금')
                self.deposit()
                continue
            elif inum == 3:
                print('선택 메뉴: 3. 출금')
                self.withdraw()        
                continue
        
    def disBalance(self):
        print(f'잔액은 {self.amount}원 입니다.')
        
    def deposit(self):
            self.process = True
            while self.process:
                self.deposit = int(input('입금할 금액을 알려주세요: '))
                if self.deposit <= 0:
                    print(f'{self.deposit}원을 입금할 수 없습니다. 정확한 금액을 알려주세요.')
                    continue
                else:
                    self.amount += self.deposit
                    print(f'{self.deposit}원 입금되었습니다. 잔액은 {self.amount}원 입니다.')
                    break
                
    def withdraw(self):
            self.process = True
            while self.process:
                self.withdraw = int(input('출금할 금액을 알려주세요: '))
                if self.withdraw > self.amount :
                    print('잔액 부족입니다. 거래가 거절되었습니다.')
                    break
                if self.withdraw <= 0:
                    print(f'{self.withdraw}원을 출금할 수 없습니다. 정확한 금액을 알려주세요.')
                    continue
                else:
                    self.amount -= self.withdraw
                    print(f'{self.withdraw}원 출금되었습니다. 잔액은 {self.amount}원 입니다.')
                    break
```

```python
#교수님 버전

class CAccount2:
    def __init__(self,owner,amount = 0,password = 1111):
        self.owner = owner
        self.password = password
        self.balance =amount
        
        message = '''
---------------
1. 잔액확인
2. 입금
3. 출금
4. 종료
---------------
'''

        self.dispTitle()
        
    def bankTransaction(self):
        initpass = int(input('계좌에 등록된 패스워드를 입력해주세요:'))    
        if initpass != self.password:
            self.dispTitleErr()
            return
        
        inum = int(input('메뉴를 선택해주세요:'))

        if inum == 4:
            self.dispEndbanking()
            return
        
        if inum == 1:
            self.dispBalance()
        elif inum == 2:
            self.deposit()
        elif inum == 3:
            self.withdraw()
        
        
    def deposit(self):
        inAmount = int(input('입금할 금액을 알려주세요: '))
        if inAmount < 0:
            self.dispAmountErr()
        else:
            self.balance += inAmount
            self.dispDepositSucc(inAmount)
            self.dispBalance()
    
    def withdraw(self):
        inAmount = int(input('출금할 금액을 알려주세요: '))
        if inAmount < 0:
            self.dispAmountErr()
        elif self.balance < inAmount:
            self.dispBalance() 
        else:
            self.balance -= inAmount
            self.dispWithdrawSucc(inAmount)
            self.dispBalance()
            
    #Display messages
    
    def dispTitle(self):
        print("%s 통장" %self.owner)
        print(message)

    def dispTitleErr(self):
        print("%s 통장" %self.owner)
        print(message)
        
    def dispBalance(self):
        print('잔액은 %d 입니다.' %self.Balance)
        
    def dispBalanceErr(self):
        print("잔액부족, 거래가 거절되었습니다.")
        
    def dispAmountErr(self):
        print('정확한 금액을 입력하세요')
        
    def dispDepositSucc(self,amt):
        print(amt,'원이 성공적으로 입금되었습니다.')   
        
    def dispWithdrawSucc(self,amt):
        print(amt,'원이 성공적으로 인출되었습니다')
        
    def dispEndbanking(self):
        print('은행 업무를 종료합니다.')
            
```
