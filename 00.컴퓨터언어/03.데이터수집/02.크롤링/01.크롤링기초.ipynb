# 연습문제

응용 : 영화관 데이터 수집

1단계: 크롤링 https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR

2단계 : 파싱 class="type_plural list_exact movie_list"

3단계 : 원하는 데이터 수집 (별점(score), 참여(review), 개봉(info_tit  dl_comm), 연간, 누적) li들 중에 다섯개만 출력


```python
# 1단계: 크롤링 https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR

import requests
import bs4

res = requests.get("https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR")
res.raise_for_status()

print(res.status_code)
print(len(res.text))
```

```python
# 2단계 : 스크래핑 class="type_plural list_exact movie_list"

html_str = str(res.text)
bs_obj = bs4.BeautifulSoup(html_str,"html.parser")

ol = bs_obj.find("ol",{"class":"list_exact"}) #찾으려는 데이터들이 모두 포함되었으면서 다른 영역과 구분된 영역
movies = ol.findAll("a",{"class":"tit_main"}) #.text를 붙이면 여기서부터 텍스트만 추출된다
 # = divs = ol.findAll("div",{"class":"info_tit"})
dls = ol.findAll("dl",{"class":"dl_comm"})
scores = ol.findAll("dd",{"class":"score"})
link_counts = ol.findAll("a",{"class":"link_count"})

print(ol)
```

```python
# 3단계 : 원하는 데이터 수집 (별점, 참여, 개봉, 연간, 누적)
# 완성본 01
for i in range(0,5):
    print(movies[i].text)

    
    if i == 0:
        x = (i+1)*4
        print(dls[x-4].text)
        print(dls[x-3].text)
        print(dls[x-2].text)
        print(dls[x-1].text)
        print()
    else:
        x = i * 4
        print(dls[x].text)
        print(dls[x+1].text)
        print(dls[x+2].text)
        print(dls[x+3].text)
        print()
```

```python
# 3단계 : 원하는 데이터 수집 (별점, 참여, 개봉, 연간, 누적)
#완성본 02

for i in range(0,5):
    print(movies[i].text)
    print(scores[i].text,":",link_counts[i].text)
    
    if i == 0:
        x = (i+1)*4
        print(dls[x-3].text)
        print(dls[x-2].text)
        print(dls[x-1].text)
        print()
    else:
        x = i * 4
        print(dls[x+1].text)
        print(dls[x+2].text)
        print(dls[x+3].text)
        print()
```

```python
# 영화제목은 0,1,2,3,4 로 스텝이 1인데 나머지는 스텝이 4라서 스텝 계산 ->
for i in range(0,5):
    print(i)
    if i == 0:
        x = (i+1)*4
        print(x-4)
        print(x-3)
        print(x-2)
        print(x-1)
        print()
    else:
        x = i * 4
        print(x)
        print(x+1)
        print(x+2)
        print(x+3)
        print()
```

### 연습문제 - 쿠팡
크롤링할 정보 : 상품명, 가격 등등

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service,options=options)

url = "https://www.coupang.com/np/categories/194276"
driver.get(url)

productList = driver.find_element(By.ID,"productList")

productNames = productList.find_elements(By.CLASS_NAME,"name")
productPrices = productList.find_elements(By.CLASS_NAME,"price-value")
productDelivery = productList.find_elements(By.TAG_NAME,"em")
productRating = productList.find_elements(By.CLASS_NAME,"rating")
productLinks = productList.find_elements(By.CLASS_NAME,"baby-product-link")  #상품링크
```

```python
#상품명

for idx, e in enumerate(productNames):
    names = e.text
    print(names)
    time.sleep(0.5)
```

```python
#가격

for idx, e in enumerate(productPrices):
    prices = e.text
    print(prices)
    time.sleep(0.5)
```

```python
#도착예정일

for idx, e in enumerate(productDelivery):
    delivery = e.text
    print(delivery)
    time.sleep(0.5)
```

```python
#평점

for idx, e in enumerate(productRating):
    rates = e.text
    print(rates)
    time.sleep(0.5)
```

```python
#상품링크

for idx, e in enumerate(productLinks):
    links = e.get_attribute("href")
    
    print(links)
    time.sleep(0.5)

```

```python
#상품이미지 

productimgs = driver.find_elements(By.CLASS_NAME,"image")
print(productimgs)
```

```python

for idx, e in enumerate(productimgs):
    srcs = e.get_attribute("src")
    
    print(srcs)
    time.sleep(0.5)
    
    f = open("image{o}".format(e),"wb")
    f.write(srcs.content)
    f.close()
```
