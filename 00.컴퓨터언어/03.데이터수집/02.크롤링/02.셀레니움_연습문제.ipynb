### 나의 얼레벌레

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.coupang.com/np/categories/194276")
productList = driver.find_element(By.ID, 'productList')
lis= productList.find_elements(By.TAG_NAME, 'li')
lis
```

```python
import csv
file_name = 'productlist.csv'
```

```python
import openpyxl

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = 'product_list'
```

```python
# 완성본

productNames = []
productPrices = []
productDeliveries = []
productRatings = []
productLinks = []

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = 'product_list'

for idx,li in enumerate(lis):
    e = li.find_element(By.TAG_NAME,"img")
    imgsrc = e.get_attribute("src")
    img = requests.get(imgsrc)
 
    with open("./02.selenium/product{0}.jpg".format(idx+1),"wb") as f:
        f.write(img.content)
        f.close()
    

    name = li.find_element(By.CLASS_NAME,"name").text
    productNames.append(name)
    price = li.find_element(By.CLASS_NAME,"price-value").text
    productPrices.append(price)
    delivery = li.find_element(By.TAG_NAME,"strong").text
    productDeliveries.append(delivery)
    rate = li.find_element(By.CLASS_NAME,"rating").text
    productRatings.append(rate)
    delivery = li.find_element(By.CLASS_NAME,"delivery").text
    productDeliveries.append(delivery)
    
    linksrc = li.find_element(By.CLASS_NAME,"baby-product-link")
    links = linksrc.get_attribute("href")

    excel_sheet.append([name, price, delivery,rate,links,f'product{idx+1}'])

    excel_file.save('product_list.xlsx')
    excel_file.close()
    
```

### 교수님 완성본

```python
# 교수님 완성본
import pandas as pd

productNames = []
productPrices = []
productDeliveries = []
productRatings = []
productLinks = []
productImages = []


for idx,li in enumerate(lis):
    try:
        e = li.find_element(By.TAG_NAME,"img")
        imgsrc = e.get_attribute("src")
        img = requests.get(imgsrc)

        image_name = "./02.selenium/product{0}.jpg".format(idx+1) 
        with open(image_name,"wb") as f:
            f.write(img.content)

        name = li.find_element(By.CLASS_NAME,"name").text
        productNames.append(name)
        price = li.find_element(By.CLASS_NAME,"price-value").text
        productPrices.append(price)
        delivery = li.find_element(By.TAG_NAME,"strong").text
        productDeliveries.append(delivery)
        rate = li.find_element(By.CLASS_NAME,"rating").text
        productRatings.append(rate)
        delivery = li.find_element(By.CLASS_NAME,"delivery").text
        if delivery :
            productDeliveries.append(delivery)
        else:    
            productDeliveries.append("-")
        linksrc = li.find_element(By.CLASS_NAME,"baby-product-link").get_attribute("href")
        productLinks.append(linksrc)
        productImages.append(image_name)

    except Exception:
        pass
    
    
final = pd.DataFrame({'이름':productNames})
final.to_csv('쿠팡식품_이름.csv',encoding='euc-kr')
final                      
```

```python

import pandas as pd
prd_name = []
prd_price = []
prd_deli = []
prd_url = []  # 상세페이지
prd_foodname = [] # 식품 이미지 이름
for idx, li in enumerate(lis):
    try:
        e= li.find_element(By.TAG_NAME,'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()

        food_name = "./02.selenium//food{}.jpg".format(idx+1)
        with open(food_name, "wb")  as f:
            f.write(img_res.content)
        prd_name.append(li.find_element(By.CLASS_NAME, 'name').text)
        prd_price.append(li.find_element(By.CLASS_NAME, 'price-value').text)
        prd_deli.append(li.find_element(By.CLASS_NAME, 'delivery').text)
        prd_url.append(li.find_element(By.CLASS_NAME, 'baby-product-link').get_attribute('href'))
        prd_foodname.append(food_name)

    except Exception:
        pass
final = pd.DataFrame({'이름': prd_name, '가격': prd_price, 'Delivery': prd_deli, 'URL': prd_url, '파일이름' : prd_foodname})
final.to_csv('쿠팡식품.csv', encoding='euc-kr')
final
```
