```python
# 페이지 각각 해보기
page = 1

url = "https://www.e-himart.co.kr/app/display/showDisplayCategory?dispNo=1011030000#pageCount={}".format(page)
driver = webdriver.Chrome(service=service,options=options)
driver.get(url)

uls = driver.find_element(By.ID,"productList")
lis = uls.find_elements(By.TAG_NAME,"li")

prdLinks=[]
prdNames = []
prdPrices = []
prdRatings = []
prdImages = []
imageNames = []
# time.sleep(sec)

try:
    for idx, li in enumerate(lis):
        imgs = li.find_element(By.TAG_NAME,"img")
        imgsrc = imgs.get_attribute("src")
        imgres = requests.get(imgsrc)
        
        imageName = "./02.Selenium/himart_page{1}_prd{0}.jpg".format(idx+1,page)
        with open(imageName,"wb") as f:
            f.write(imgres.content)
            
        prdNames.append(li.find_element(By.CLASS_NAME,"prdName").text)
        prdPrices.append(li.find_element(By.CLASS_NAME,"discountPrice").text)
        prdLinks.append(li.find_element(By.TAG_NAME,"a").get_attribute("href"))
        try:
            rating = li.find_element(By.CLASS_NAME,"ratingPoint").text
            prdRatings.append(rating)        
        except:
            prdRatings.append("-")    
        time.sleep(0.2)
except Exception as e:  
    pass
```

```python
import pandas as pd

final = pd.DataFrame({'페이지':page,'이름':prdNames,'가격':prdPrices,'별점':prdRatings,'링크':prdLinks})
final.to_csv('하이마트_냉장고.csv',encoding='euc-kr')
final  
```

```python
# 5페이지까지 한번에 가기 
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
import time
import pandas as pd

service = Service()
options = webdriver.ChromeOptions()

prdLinks=[]
prdNames = []
prdPrices = []
prdDelivery = []
prdRatings = []
prdImages = []
imageNames = []
# time.sleep(sec)
pages = []

for i in range(0,5):
    url = "https://www.e-himart.co.kr/app/display/showDisplayCategory?dispNo=1011030000#pageCount={}".format(i+1)
    driver.get(url)
    top = driver.find_element(By.CLASS_NAME,"cateWrap")
    # uls = top.find_element(By.CLASS_NAME,"thumnailS")
    uls = top.find_element(By.ID,"productList")
    lis = uls.find_elements(By.TAG_NAME,"li")
    page = i + 1
    for idx, li in enumerate(lis):
        try:
            prdImages.append(li.find_element(By.TAG_NAME,"img"))
            try:
            # pages.append(page)
                imgs = li.find_element(By.TAG_NAME,"img")
                imgsrc = imgs.get_attribute("src")
                imgres = requests.get(imgsrc) 
                imageName = "./02.Selenium_himart/page{1}_prd{0}.jpg".format(idx+1,page)
                with open(imageName,"wb") as f:
                    f.write(imgres.content)
                imageNames.append(imageName)            
                time.sleep(0.2)
            except:
                imageNames.append("-")
        except:
            imageNames.append("-")
            prdImages.append("_")
        try:
            prdNames.append(li.find_element(By.CLASS_NAME,"prdName").text)
        except:
            prdNames.append("-")
        
        prdLinks.append(li.find_element(By.TAG_NAME,"a").get_attribute("href"))
        #price
        try:
            prices = li.find_elements(By.CLASS_NAME,"discountPrice")
            if len(prices) % 2 == 0:
                prdPrices.append(prices[1].text)
            else:
                prdPrices.append(prices[0].text)
        except:
            prdPrices.append("-")
        #delivery
        try:
            prdDelivery.append(li.find_element(By.CLASS_NAME,"delivery").text)
        except:
            prdDelivery.append("-")
        #rating
        try:
            prdRatings.append(li.find_element(By.CLASS_NAME,"ratingPoint").text)        
        except:
            prdRatings.append("-")    
        time.sleep(0.2)
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
prd_name = []
prd_price = []
prd_deli = []
prd_star = []
prd_url = []
prd_imgname = []
prd_name = []
prd_price = []
prd_deli = []
prd_star = []
prd_url = []
prd_imgname = []
for page in range(1, 6):
    print("*** page ***", page)
    url = f'http://www.e-himart.co.kr/app/display/showDisplayCategory?dispNo=1011030000#pageCount={page}'
    driver.get(url)
    cateWrap = driver.find_element(By.CLASS_NAME, 'cateWrap')
    productList = cateWrap.find_elements(By.ID, 'productList')
    lis= productList[0].find_elements(By.TAG_NAME, 'li')
    for idx, li in enumerate(lis):
        try:
            prd_name.append(li.find_element(By.CLASS_NAME, 'prdName').text)
        except:
            prd_name.append("-")
        try:
            tmpDiscountPrice = li.find_elements(By.CLASS_NAME, 'discountPrice')
            if(len(tmpDiscountPrice) == 2):
                prd_price.append(li.find_elements(By.CLASS_NAME, 'discountPrice')[1].text)
            else:
                prd_price.append(li.find_elements(By.CLASS_NAME, 'discountPrice')[0].text)
        except:
            prd_price.append("-")
        try:
            prd_deli.append(li.find_element(By.CLASS_NAME, 'delivery').text)
        except:
            prd_deli.append("-")
        try:
            prd_star.append(li.find_element(By.CLASS_NAME, 'ratingStar').text)
        except:
            prd_star.append("-")
        try:
            prd_url.append(li.find_element(By.TAG_NAME, "a").get_attribute('href'))
            try:
                e= li.find_element(By.TAG_NAME,'img')
                img_url = e.get_attribute('src')
                img_res = requests.get(img_url)
                # img_res.raise_for_status()
                image_name = "./images-himart/image{}.jpg".format(idx+1)
                with open(image_name, "wb")  as f:
                    f.write(img_res.content)
                prd_imgname.append(image_name)
            except:
                prd_imgname.append("-")
        except:
            prd_url.append("-")
            prd_imgname.append("-")
        if idx >= 5:
            break
print(len(prd_name), len(prd_price), len(prd_deli), len(prd_star), len(prd_url), len(prd_imgname))
final = pd.DataFrame({'이름': prd_name, '가격': prd_price, 'Delivery': prd_deli, '별점' : prd_star, 'URL': prd_url, \
    'image':prd_imgname})
final.to_csv('하이마트.csv', encoding='euc-kr')
final
```
