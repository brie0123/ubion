# matplotlib

## 데이터 전처리

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../05.데이터분석/00.data/population_city.xlsx',header=0)
```

```python
# .fillna 함수 : 엑셀 시트에서 병합되있는 애들 데이터 채우기
df = df.fillna(method='ffill')
df.head()
```

```python
# mask 연산을 활용한 전출지가 서울특별시인 데이터만 추출
mask = (df['전출지별'] == '서울특별시') # mask = (조건문)이 True인 값만 추출
mask.tail()
```

```python
df_seoul = df[mask] 
df_seoul
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'],axis=1)
df_seoul.rename({'전입지별':'전입지'},axis=1, inplace=True)
df_seoul.set_index('전입지',inplace=True)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
print(sr_one)
```

## 시각화 단계

```python
import matplotlib.pyplot as plt
plt.plot(sr_one.index,sr_one.values)
plt.show()
```

```python
plt.plot(sr_one)
plt.show()
```

```python
# 내장폰트가 아닌 외부 폰트 적용하기
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
# font_path = './00.data/malgun.ttf'
# font_name = font_manager.FontProperties(frame=font_path).get_name()
rc('font',family='paybooc')
plt.figure(figsize=(14,5))

plt.xticks(rotation='vertical')

plt.plot(sr_one,marker='o',markersize=10)
plt.title('서울 -> 경기 인구 이동')

plt.xlabel('기간')
plt.ylabel('이동 인구 수')
plt.legend(labels=['서울 -> 경기'],loc=7,fontsize=15) # .legend(): 범례 표시

#y축 범위 지정(최솟값, 최댓값)
plt.ylim(5000,800000)

#주석 표시 - 화살표
plt.annotate("",
             xy=(20,620000),                #화살표 머리 부분(끝점)
             xytext=(2,290000),             #화살표 꼬리 부분(시작점)
             xycoords='data',               #좌표체계
             arrowprops=dict(arrowstyle="->",color='darkseagreen',lw=5),)    #화살표 서식

# 주석 표시 - 텍스트
plt.annotate('',
            xy=(47, 450000),       #화살표의 머리 부분(끝점)
            xytext=(30, 580000),   #화살표의 꼬리 부분(시작점)
            xycoords='data',       #좌표체계
            arrowprops=dict(arrowstyle='->', color='olive', lw=5),  #화살표 서식
            )


plt.annotate('인구이동 증가(1970-1995)',        #텍스트 입력
             xy=(10,550000),                #텍스트 위치 기준점
             rotation=25,                   #텍스트 회전각도
             va='baseline',                 #텍스트 상하 정렬
             ha='center',                   #텍스트 좌우 정렬
             fontsize=15,)                  #텍스트 크기

plt.annotate('인구이동 감소(1995-2017)',        #텍스트 입력
             xy=(40,560000),                #텍스트 위치 기준점
             rotation=-11,                  #텍스트 회전각도
             va='baseline',                 #텍스트 상하 정렬
             ha='center',                   #텍스트 좌우 정렬
             fontsize=15,)                  #텍스트 크기

plt.show()
```

## 데이터 살펴보기

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

print(titanic.head())
print('\n')
print(titanic.info())
```

## 데이터 시각화

```python
# style 옵션 : darkgrid, whitegrid, dark, white, ticks



sns.set_style('darkgrid')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)  # .add_subplot() : 차트 두 개로 하기
ax2 = fig.add_subplot(1,2,2)

# 선형회귀선 표기 (fit_reg = True)
sns.regplot(x='age', y='fare', data=titanic, ax=ax1)  # .regplot = 서로 다른 두 변수 사이의 상관관계(?)를 산점도로 보여주고, 선형 회귀선을 표시할 수 있는 그래프 타입

# 선형회귀선 미표기 (fit_reg = False)
sns.regplot(x='age',y='fare',data=titanic,ax=ax2,fit_reg=False)
```

### 단변수 그래프

#### .distplot() / .kdeplot() / .histplot()

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

# 그래프 객체 생성( figure에 3개의 서브플롯 생성)

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# .displot
sns.distplot(titanic['fare'],ax=ax1)

# .kdeplot
sns.kdeplot(x='fare',data=titanic,ax=ax2)

# .histplot
sns.histplot(x='fare',data=titanic,ax=ax3)

#차트 제목 표시
ax1.set_title('titanic fare - distplot')
ax2.set_title('titanic fare - kdeplot')
ax3.set_title('titanic fare - histplot')

plt.show()
```

### 다중변수 그래프

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('white')

table = titanic.pivot_table(index=['sex'],columns=['class'],aggfunc=['size'])
```

#### .heatpmap() : 히트맵

```python
# 히트맵 그리기
sns.heatmap(table, annot = True, fmt = 'd', cmap = 'YlGnBu',linewidth=5,cbar=False)

plt.show()
```

```python
# 히트맵 그리기
sns.heatmap(table, annot = True, fmt = 'd', cmap = 'YlGnBu',linewidth=5,cbar=True)
# 속성 : annot : 안에 숫자 표기 or not, cbar = 컬러바
print(table)
plt.show()
```

#### .swarmplot() / .stripplot()

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('white')
```

```python
# 범주형(categorical) 변수의 값 확인
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
```

```python
# 이산형 변수의 분포 = 데이터 분산 고려 X
sns.stripplot(x='class',y='age',data=titanic,ax=ax1)

# 이산형 변수의 분포 - 데이터 분산 고려 O
sns.swarmplot(x='class',y='age',data=titanic,ax=ax2)

ax1.set_title('Strip plot')
ax2.set_title('Swarm plot')
plt.show()
```

#### .barplot() : 막대그래프

```python
#

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.barplot(x='sex',y='survived',data=titanic,ax=ax1)
sns.barplot(x='sex',y='survived',hue='class',data=titanic,ax=ax2)   # hue = 하위 그룹화 같은건강
sns.barplot(x='sex',y='survived',hue='class',dodge=False,data=titanic,ax=ax3)  # dodge 누적 ()

ax1.set_title('titanic survived - sex')
ax2.set_title('titanic survived - sex/class')
ax3.set_title('titanic survived - sex/class(stacked)')

```

#### .countplot()

```python

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.countplot(x='class',palette='Set1',data=titanic, ax=ax1)
sns.countplot(x='class',hue = 'who',palette='Set2',data=titanic, dodge=True,ax=ax2)
sns.countplot(x='class',hue = 'who',palette='Set3',data=titanic, dodge=False, ax=ax3)

ax1.set_title('titanic survived')
ax2.set_title('titanic survived - who')
ax3.set_title('titanic survived - who(stacked)')

```

#### .violinplot() : 바이올린그래프

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

fig=plt.figure(figsize=(15,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

# 박스그래프 - 기본값
sns.boxplot(x = 'alive',y = 'age', data=titanic, ax=ax1)
sns.boxplot(x = 'alive',y = 'age', hue='sex', data=titanic, ax=ax2)
sns.violinplot(x = 'alive',y = 'age', data=titanic, ax=ax3)
sns.violinplot(x = 'alive',y = 'age', hue='sex', data=titanic, ax=ax4)

plt.show()
```

```python
from folium.map import Popup
import pandas as pd
import folium

unidata = pd.read_excel('./01.map/서울지역 대학교 위치.xlsx')
seoul_map = folium.Map(location=[37.55,126.98], zoom_start=12)

for name, lat, lng in zip(unidata.이름, unidata.위도, unidata.경도):
    iframe = folium.IFrame(name,width=150, height = 50)
    popup = folium.Popup(iframe, max_width = 600)
    folium.Marker([lat,lng],popup=popup).add_to(seoul_map)
    
seoul_map.save('./01.map/seoul_colleges.html')
seoul_map
```
