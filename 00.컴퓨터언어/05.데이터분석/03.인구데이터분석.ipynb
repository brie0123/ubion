#### 지역명 검색해서 그래프 띄워주기

```python
import csv

f = open("./00.data/age.csv","r",encoding='cp949')
data = csv.reader(f,delimiter=',')
print(data)

result=[]
region = input('인구 수를 알고 싶은 지역명(읍면동 단위)을 입력하세요: ')

try:
    for row in data:
        if region in row[0]:
            for i in row[3:]:    
                result.append(int(i))
except:
    print('읍면동 이름을 입력해주세요')

f.close()

import matplotlib.pyplot as plt

plt.plot(result)
plt.show()
```

### 성별에 따른 인구 수

```python
import csv

f = open("./00.data/gender.csv","r",encoding='cp949')
data = csv.reader(f,delimiter=',')
print(data)

male = []
female = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            male.append(-int(row[i+3]))
            female.append(int(row[-(i+1)]))

female.reverse()
f.close()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font',family="paybooc")  # font 저장된 이름 찾기 -> 파인더 - 라이브러리 - 폰트
plt.title('신도림 성별에 따른 인구 수')
plt.barh(range(101),male,label='male',color='thistle')
plt.barh(range(101),female,label='female',color='bisque')
plt.legend(loc=2)
plt.show()
```

#### 지역 검색해서 성별 비율 파이로 그리기

```python
#데이터 뽑기
import csv

f = open("./00.data/gender.csv","r",encoding='cp949')
data = csv.reader(f,delimiter=',')
print(data)

sum_male = 0
sum_female = 0
region = input('거주자 성별 비율을 알고 싶은 지역을 입력하세요: ')
# village = input('정확한 읍면동을 입력해주세요: ')

for row in data:
    search = row[0].split()
    if region in row[0]:
        target = str(row[0].strip())
        for i in row[3:104]:
            sum_male += int(i)
        for i in row[106:]:
            sum_female += int(i)
        break #도 혹은 시 단위의 지역을 입력하면 row[0]를 걸러내는 조건문에서 row가 중복되기 때문에 가장 상위 지역의 데이터를 찾기 위해 break를 건다
f.close()

size = []
size.append(sum_male)
size.append(sum_female)
```

```python
#차트그리기
import matplotlib.pyplot as plt
# size = [sum_male,sum_female]
plt.rc('font',family='paybooc')
label = ['male','female']
color = ['olivedrab','darkseagreen']
plt.title(target+'지역 거주자 성비')
plt.axis('equal')
plt.pie(size,labels=label,autopct = '%.2f%%',colors = color,startangle = 90)  # startangle = 비율을 그리는 시작점 지정. 뭔가 파이차트는 반시계방향으로 데이터를 그리는 듯
plt.legend()
plt.show()
```

## 꺾은선 그래프 그리기

```python
#데이터 뽑기
import csv

f = open("./00.data/gender.csv","r",encoding='cp949')
data = csv.reader(f,delimiter=',')
print(data)

male = []
female = []
region = input('인구 수를 알고 싶은 지역을 입력하세요: ')
# village = input('정확한 읍면동을 입력해주세요: ')

for row in data:
    search = row[0].split()
    if region in row[0]:
        target = str(row[0].strip())
        for i in row[3:104]:
            male.append(int(i))
        for i in row[106:]:
            female.append(int(i))
        break #도 혹은 시 단위의 지역을 입력하면 row[0]를 걸러내는 조건문에서 row가 중복되기 때문에 가장 상위 지역의 데이터를 찾기 위해 break를 건다
f.close()
```

## 남자 수 - 여성 수 뺀 차이 막대그래프 그리기

```python
# 데이터 뽑기
import csv

f = open("./00.data/gender.csv","r",encoding='cp949')
data = csv.reader(f,delimiter=',')
print(data)

male = []
female = []
region = input('인구 수를 알고 싶은 지역을 입력하세요: ')
# village = input('정확한 읍면동을 입력해주세요: ')
sub_result = []
for row in data:
    search = row[0].split()[0]
    if region in row[0]:
        target = str(row[0].strip())
        for i in range(0,101):
            sub_result.append(int(row[i+3])-int(row[i+106]))
        break
f.close()
```

### 데이터 값 랜덤으로 지정해서 산점도그래프 그리기

```python
import random

x=[]
y=[]
size=[]
for i in range(100):
    x.append(random.randint(500,1000))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x,y,s=size)
plt.show()
```

```python
plt.scatter(x,y,s=size,c=range(100),cmap='YlGnBu')
plt.colorbar()
plt.show()
```

```python
# alpha= %f: 점들 투명도 설정
plt.scatter(x,y,s=size,c=range(100),cmap='YlGnBu',alpha=0.6)
plt.colorbar()
plt.show()
```

### 연습문제 : 특정 지역 연령별 데이터 산점도로 표현하기

```python
import csv

f = open("./00.data/gender.csv","r",encoding='cp949')
data = csv.reader(f,delimiter=',')
print(data)

male = []
female = []
region = input('인구 수를 알고 싶은 지역을 입력하세요: ')
# village = input('정확한 읍면동을 입력해주세요: ')

for row in data:
    search = row[0].split()[0]
    if region in row[0]:
        target = str(row[0].strip())
        for i in range(0,101):
            male.append(int(row[i+3]))
            female.append(int(row[i+106]))

        break #도 혹은 시 단위의 지역을 입력하면 row[0]를 걸러내는 조건문에서 row가 중복되기 때문에 가장 상위 지역의 데이터를 찾기 위해 break를 건다
f.close()

# 지역 남녀 산점도
```

```python

plt.figure(figsize=(10,5),dpi=100)
plt.scatter(male,female,c=range(101),cmap='BuPu',alpha=0.7)
plt.title(target+'지역의 성별 분포도')
plt.colorbar()
plt.plot(range(max(male)),range(max(female)),'g')
plt.xlabel('남성 인구 수')
plt.ylabel('여성 인구 수')

plt.show()
```
