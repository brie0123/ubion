# 딥러닝

### 라이브러리 준비

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

## 데이터 준비

```python
Data_set = np.loadtxt('./data/ThoraricSurgery.csv',delimiter=',')
Data_set
```

```python
# 환자의 기록과 수술 결과를 X와 y로 구분하여 저장

X = Data_set[:,0:17]
y = Data_set[:,17]
```

```python
# 모델 알고리즘 준비
model = Sequential()

# 모델 레이어 설정
model.add(Dense(30, input_dim=17, activation='relu'))       # 은닉층 노드 30개로 설정 -> 해당 레이어 사이는 relu함수 적용 / input_dim=n으로 최초 노드 개수 설정
model.add(Dense(1,activation='sigmoid'))                    # 출력층 설정 -> 출력층에는 sigmoid 적용된 결과 출력됨


# 학습 방식 설정
model.compile(metrics=['accuracy'],optimizer='adam',loss='binary_crossentropy')     # 이진분류이기 때문에 loss를 binary_crossentropy로 설정, 다중분류일 경우 categorical_crossentropy로 설정

# 모델 학습 
model.fit(X,y,epochs=100,batch_size=10)      # epoch = 계산해보는 w의 개수, barch_size = 데이터를 밀어넣는 단위
```
