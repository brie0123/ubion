# 교수님 버전

```python
# from tensorflow.keras.models import 
import numpy as np
import tensorflow as tf
```

```python
# 시드 고정             -> 위에서 빠진 부분

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
df = pd.read_csv('./data/sonar.csv',header=None)
```

```python
print(type(df.values))
dataset = df.values
X = dataset[:,:60].astype(float)
y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(y_obj)
Y = e.transform(y_obj)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, random_state=seed)
```

```python
model = Sequential()

# 활성함수 relu: 기울기 소실(Vanishing Gradient) 문제가 발생하지 않음, 기존 활성함수에 비해 속도가 빠름(탄젠트 함수 대비 6배 빠름)
# 최적화함수 adam: 가중치와 bias를 자동으로 조절해줌
model.add(Dense(24, input_dim=60,activation='relu'))        # hidden layer 1
model.add(Dense(10, activation='relu'))                     # hidden layer 2
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=130, batch_size=5)

# model.save('my_model.h5')       # 모델 저장 / 현재(20231219) 호환 에러로 주석처리
# H5 형식: Tensorflow, Keras를 사용하는 딥러닝 모델에서 사용, 모델구조/가중치/학습 구성/옵티마이저 상태 등을 저장 
# -> 저장한 모델은 나중에 같은 상태로 복원 가능

import pickle
pickle.dump(model, open('my_model.pkl','wb'))
```

```python
# 모델 테스트
del model
model = pickle.load(open('my_model.pkl','rb'))

print('/n Accuracy: % .4f' % (model.evaluate(X_test,y_test)[1]))
```
