```python
import pandas as pd

word_count = pd.read_csv('./data/word_count.csv')
word_count.head()
```

### 단어 구름

#### 라이브러리 호출 및 프레임 설정

```python
from wordcloud import WordCloud

wc = WordCloud(background_color='white', max_words=100, width=400, height=300,colormap='PuBu')
wc
```

#### 단어 빈도 딕셔너리 변환

```python
word_count.set_index('단어')
```

* to_dict

```python
count_dict = word_count.set_index('단어')['빈도'].to_dict()
count_dict
```

* zip 활용

```python
dict(zip(word_count['단어'],word_count['빈도']))
```

#### wc 그리기

```python
cloud = wc.fit_words(count_dict)
cloud.to_image()
```

```python
cloud.to_file('data/cloud.png')
```
