### 크롤링 대상의 구체적인 element를 정리해보자

* 크롤링할 것을 생각해보자. 1. 지역별
계층 1: 시도별
계층 2: 스타벅스 지점(서울시) 정보 : 지점명, 매장유형(일반/리저브/DT),건물정보(도로명주소,건물명,동 이름),입점 층/호수 정보 / 상세정보보기를 클릭해서 영업시간 / 위도경도


```python
# 검색할 지역 필터링하기
    # 시도 선택 : div class="loca_step1" 안에서 class="sido_arae_box"
    # 그 안에 li들,  class="set_sido_cd_btn"로 검색할 지역 선택
    # 시도 아래 구/군 선택: class="loca_step2_cont" -> findelement로 찾기(첫번째 element가 항상 전체): class="set_gugun_cd_btn"

# 검색 결과

    # 시도별 매장 검색 결과가 담긴 div : class="result_list_wrap"
    # 그 안에서 제일 하위 div : id="mCSB_3_container"
    # 매장 정보를 담고 있는 li : class="quickResultLstCon"
    
    # li 안에서
        # 위도: lat=
        # 경도: data-long=
        # 매장 이름: data-name=
        # 매장 상세정보 : class="result_details"
        # * 사이렌오더 여부: strong 태그, my_siren_order_store_yn='Y/N'
        
```

### 크롤링 시작

#### 셀레니움 크롤링

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service,options=options)

url = 'https://www.starbucks.co.kr/store/store_map.do?disp=locale'
driver.get(url)
```

```python
# 1단계 
# 검색할 지역 필터링하기
    # 시도 선택 : div class="loca_step1" 안에서 class="sido_arae_box"
    # 그 안에 li들,  class="set_sido_cd_btn"로 검색할 지역 선택
    # 시도 아래 구/군 선택: class="loca_step2_cont" -> findelement로 찾기(첫번째 element가 항상 전체): class="set_gugun_cd_btn"

time.sleep(0.3)         
# city_area = driver.find_element(By.CLASS_NAME,"sido_arae_box")
city = driver.find_element(By.CLASS_NAME,"set_sido_cd_btn")
city.click()

```

```python

region = driver.find_element(By.CLASS_NAME,"gugun_arae_box")
region = region.find_element(By.CLASS_NAME,"set_gugun_cd_btn")
region.click()
```

```python
# 검색 결과
    # 시도별 매장 검색 결과가 담긴 div : class="result_list_wrap"
    # 그 안에서 제일 하위 div : id="mCSB_3_container"
    # 매장 정보를 담고 있는 li : class="quickResultLstCon"

```

```python
result = driver.find_element(By.ID,"mCSB_3_container")
lis = result.find_elements(By.TAG_NAME,"li")
print(lis)
```

##### selenium_ 데이터가 일부만 나오는 문제

```python
store_name = []
store_lat = []
store_long = []
store_loc = []

for li in lis:
    store_name.append(li.find_element(By.TAG_NAME,"strong").get_attribute('data-name'))
    store_lat.append(li.get_attribute('data-lat'))
    store_long.append(li.get_attribute('data-long'))
    store_loc.append(li.find_element(By.CLASS_NAME,"result_details").text)
    time.sleep(0.02)
    
```

```python
print(len(store_name),len(store_loc),len(store_lat),len(store_long))
```

```python
import pandas as pd
final = pd.DataFrame({'매장명': store_name, '매장주소': store_loc, '경도': store_lat, '위도': store_long})
final.to_excel('./starbucks_sel.xlsx')
final['매장주소']
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service,options=options)

url = 'https://www.starbucks.co.kr/store/store_map.do?disp=locale'
driver.get(url)
# try:
#     webDriverwait(driver,20).until(EC.element_to_be_clickable\(By.CSS_SELECTOR,seoul_btn)).click()
# except TimeoutError:
#     print('해당 페이지가 20초 내에 열리지 않았음')
```

```python
seoul_btn = "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a"
By.CSS_SELECTOR,seoul_btn.click()
```

```python
# 스크래핑
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')

starbucks_soup_list = soup.select('li.quickResultLstCon')
print(len(starbucks_soup_list))
```

```python
# 얘네 합쳐서 for 문 돌리기
starbucks_list = []

for starbucks_store in starbucks_soup_list:
    name = starbucks_store.select('strong')[0].text.strip()
    lat = starbucks_store['data-lat'].strip()
    lng = starbucks_store['data-long'].strip()
    store_type = starbucks_store.select('i')[0]['class'][0]
    add_tel = str(starbucks_store.select('p.result_details')[0]).split('<br/>')
    address = add_tel[0].split('>')[1]
    tel = add_tel[-1].split('<')[0]
    
    starbucks_list.append([name,lat,lng,store_type,address,tel]) 
    
```

```python
import pandas as pd

columns = ['매장명','위도','경도','매장타입','주소','전화번호']
seoul_starbucks_df = pd.DataFrame(starbucks_list,columns = columns)
seoul_starbucks_df.head()
```

```python
seoul_starbucks_df.to_excel('./seoul_starbucks.xlsx')
```
