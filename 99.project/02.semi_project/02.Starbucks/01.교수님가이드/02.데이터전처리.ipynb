### 가운데 세글자를 가져오는 방법 세가지

```python
# 버전 1 -> for 문 돌리기 (지금까지 가장 익숙했던 것)
new_dong = []
for i in range(len(seoul_st['주소'])):
    new_dong.append(seoul_st['주소'][i][6:9])
```

```python
seoul_st['동이름'] = new_dong
seoul_st.sample(5)
```

```python
# 버전 2 -> 함수 만들어서 apply 활용하기
def dong (x):
    return x[6:9]

seoul_st['test'] = seoul_st['주소'].apply(dong)
seoul_st.sample(5)
```

```python
# 소정 버전
def cutting(x) :
    return x.split(' ')[1]
seoul_st['시군구별']=seoul_st['주소'].apply(cutting)
seoul_st
```

```python
# 위에서 중복 데이터를 세 칼럼이나 가지고 있으므로 칼럼 두 개 지워주기
seoul_st.drop('동이름',axis=1,inplace=True)
seoul_st.drop('test',axis=1,inplace=True)
seoul_st.head()
```

```python
# 깔쌈하게 정리된 seoul_st 데이터프레임 저장
seoul_st.to_excel('./01.데이터수집/data_final/seoul_st_store_info.xlsx',index=False)
```

## 01-2.스타벅스 분석 데이터 만들기
* 시군구 목록 데이터 + 시군구별 스타벅스 매장 수 + 주민인구 수 + 사업체 수

```python
# 위에서 깔쌈하게 정리한 스타벅스 매장 목록 데이터 불러오기

seoul_st = pd.read_excel('./01.데이터수집/data_final/seoul_st_store_info.xlsx')
seoul_st.head()
```

### 구 기준으로 스타벅스 매장 수 세기

```python
# 시군구별 스타벅스 매장 수 세기
seoul_gus = []
for gu in seoul_st['시군구별'].unique():
    seoul_gus.append(gu)
seoul_gus
```

```python
seoul_st['시군구별']
```

```python
# 위에가 뻘짓이라고 생각했으나 타입이 다르다 이 두 차이는 무슨 차이인지 잘 모르겠다.
print(type(seoul_gus),type(seoul_st['시군구별']))
```

```python
# 이상하게 수 세기
gus_count = []
gus_index=[]
for i in range(len(seoul_gus)):
    condition = seoul_st['시군구별'] == seoul_gus[i]
    gus_count.append(len(seoul_st[condition]))
    gus_index.append(i)

# 제대로 되었는지 확인
# print(gus_count)
# print(gus_index)

# 위에서 만든 리스트들 데이터프레임으로 만들기
test_df = pd.DataFrame(gus_index)
test_df

# 칼럼추가
test_df['시군구별'] = seoul_gus
test_df['매장수'] = gus_count
test_df

# 인덱스 정리
# test_df = test_df.set_index('gus_index')

#완성
test_df
```

### 시군구 목록 데이터 + 스타벅스 매장 수 병합

```python
# 시군구 목록 데이터 준비
seoul_sgg = pd.read_excel('./01.데이터수집/seoul_sgg_list.xlsx')
seoul_sgg.sample(5)
```

```python
# 서울시 시군구 목록 데이터에 스타벅스 매장 수 데이터를 병합 -> 그냥 칼럼 정렬 같이 하고 더하기 *칼럼 정렬에 따라 부정확한 데이터 발생 가능

seoul_sgg.sample(5)
seoul_sgg = seoul_sgg.sort_values(['시군구명'])
seoul_sgg.reset_index()
seoul_sgg.head()

seoul_sgg['스타벅스 매장수'] = gus_count
seoul_sgg.head()
```

#### merge를 사용해서 두 데이터프레임을 데이터오염 없이 결합하자

```python
# 위에서 정리해둔 test_df 준비
test_df.head()
# test_df.drop(0,axis=1,inplace=True)
# test_df.head()
```

```python
# 키값이 될 칼럼 이름이 다름 -> test_df '시군구별' 칼럼 이름 정리해주기
test_df.rename(columns={'시군구별':'시군구명'},inplace=True)
test_df.head()
```

```python
# 시군구 목록에 시군구기준 스타벅스 매장 수 카운트 칼럼 결합 완료
merge_left = pd.merge(seoul_sgg,test_df,how='left',on='시군구명')
merge_left.head()
```

#### 교수님 _ pivot_table 사용

```python
starbucks_st_count = seoul_st.pivot_table(index = '시군구별',values='매장명',aggfunc = 'count').rename(columns={'매장명':'스타벅스_매장수'})
```

```python
starbucks_st_count
```

```python
seoul_sgg2 = pd.merge(seoul_sgg,starbucks_st_count,how='left',left_on='시군구명',right_on='시군구별')
seoul_sgg2.head()
```

```python
# 나의 수제 얼레벌레 데이터프레임이랑 비교
merge_left.head()
```

## 01-3. 시군구 기준 서울시 인구도 같이 넣어주기

```python
# 데이터 준비
seoul_sgg_pop = pd.read_excel('./01.데이터수집/seoul_pop.xlsx')
seoul_sgg_pop.head()
```

```python
# left_on, right_on을 사용할 경우 칼럼이 중복되니 가능한 on을 쓰자
seoul_st_final = pd.merge(seoul_sgg2,seoul_sgg_pop,how = 'left', left_on = '시군구명',right_on = '시군구별')
seoul_st_final.head()
```

```python
# 칼럼 정리
seoul_st_final.drop('시군구별',axis=1,inplace=True)
seoul_st_final.head()
```

```python
seoul_st_final.to_excel('./01.데이터수집/data_final/seoul_st_pop.xlsx',index=False)
```

## 01-4. 시군구 기준 서울시 산업체, 종사자 현황 추가해주기

```python
# 사업체 수 및 종사자 수 데이터 준비
seoul_biz_df = pd.read_excel('./01.데이터수집/seoul_biz.xlsx')
seoul_biz_df.head()
```

```python
seoul_st_final = pd.merge(seoul_st_final,seoul_biz_df,how = 'left', on = '시군구명')
seoul_st_final.head()
```

```python
seoul_st_final.to_excel('./01.데이터수집/data_final/seoul_st_pop_biz.xlsx',index=False)
```
