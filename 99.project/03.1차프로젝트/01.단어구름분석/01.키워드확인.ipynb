```python
df = pd.read_excel('./data/hwahae_reviews.xlsx')
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df
```

```python
def keyword_check(df,keyword,brand):
    df.dropna(subset="content",axis=0,inplace=True)
    filtered_df = df[df['content'].str.contains(keyword)]
    filtered_df.to_excel(f'./data/{brand}_keywordcheck.xlsx',index=False)
    return filtered_df
```

```python
hawhae = keyword_check(df,'리뷰','hawhae')
hawhae = hawhae.sort_values('score',ascending=False)
hawhae
```
