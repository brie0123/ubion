```python
def info_prep(df):
    df['용량'] = '-'
    df['유통기한'] = '-'
    df['성분'] = '-'
    for idx in range(len(df['상품정보'])):
        for info in df.loc[idx,'상품정보'].split("'"):
            if '용량' in info:
                splits = info.split(" ")
                for split in splits:
                    if 'ml' in split:
                        df.loc[idx,'용량'] = split.strip()
            if '사용기한' in info:
                df.loc[idx,'유통기한'] = info.split(") ")[1].strip()
            if '성분' in info:
                ingredient = info.split('성분 ')[-1].strip()
                ingredient = ingredient.replace('"',"")
                ingredient = ingredient.replace("@",",")
                df.loc[idx,'성분'] = ingredient
    return df       


def name_prep(df):
    for idx,prdname in enumerate(df['제품명']):
        if "]" in prdname:
            prdname = prdname.replace("]", "] ")
            prdname = prdname.split("] ")[1].strip()
        if "ml" in prdname:
            splits = prdname.split(" ")
            num = [splits.index(split) for split in splits if 'ml' in split]
            df.loc[idx, '제품명'] = " ".join(splits[1:num[0]]).strip()
    return df


def drop_unnorm(df):
    df['성분'] = df['성분'].apply(lambda x: x.replace("■ ","").strip())
    df['유통기한'] = df['유통기한'].apply(lambda x: x.replace("■ ","").strip())
    df = df.dropna(axis=0)
    filtered_df = df[(df["성분"].str.contains("\[")) | (df['성분'] == "-") | (df['성분'] == "")]
    target_index = list(filtered_df.index)
    df = df.drop(target_index,axis=0)
    df = df.drop_duplicates(subset="제품명")
    return df

def prep(df):
    df = info_prep(df)
    df = name_prep(df)
    df = drop_unnorm(df)
    return df
```

```python
prep_df = prep(df)
# 세럼
df = info_prep(df)
df = name_prep(df)
prep_df = df.dropna(axis=0)
# print(prep_df.info())
prep_df
```

```python
target = 'serum' # 세럼
prep_df.to_csv(f'./output/prep_{target}.csv',index=False)
```

```python
def categorized(df,category):
    df['제품유형'] = f'{category}'
    return df
```

```python
import pymysql

def update_prd_info(df):
    df = df.dropna()
    df = df.drop_duplicates()
    df = df[['제품명','브랜드', '가격', '용량', '성분', '유통기한', '제품유형']]
    # Database connection
    conn = pymysql.connect(host='localhost', user='root', password='apple123!!', db='beaulol', charset='utf8')

    with conn.cursor() as curs:
        # Create table if it does not exist
        curs.execute("""
        CREATE TABLE IF NOT EXISTS product_info (
            prd_name VARCHAR(500),
            prd_brand VARCHAR(50),
            prd_price VARCHAR(10),
            prd_capa TEXT,
            prd_ingredients TEXT,
            prd_exp TEXT,
            prd_category VARCHAR(10),
            PRIMARY KEY (prd_name)
        );
        """)

        # Prepare the SQL statement for inserting data
        sql = """
        REPLACE INTO product_info (prd_name, prd_brand, prd_price, prd_capa, prd_ingredients, prd_exp, prd_category)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        """

        # Convert DataFrame to a list of tuples
        data_tuples = [tuple(x) for x in df.to_numpy()]

        # Execute the SQL statement in bulk
        curs.executemany(sql, data_tuples)

    # Commit changes and close connection
    conn.commit()
    conn.close()
```

```python
update_prd_info(cream)
update_prd_info(lotion)
update_prd_info(serum)
update_prd_info(skin)
```
