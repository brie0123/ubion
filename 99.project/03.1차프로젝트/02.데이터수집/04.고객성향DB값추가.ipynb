```python
def generate_fruit_combinations(fruits, max_fruits=3):
    num_fruits = np.random.randint(1, max_fruits + 1)  # Random number of fruits
    return ', '.join(np.random.choice(fruits, size=num_fruits, replace=False))

# Generate the 'Fruit' column with multiple values per cell
# multiple_fruits = [generate_fruit_combinations(serum,5) for value in range(serum)]
multiple_serum = [generate_fruit_combinations(serum,5) for _ in range(num_samples)]

# Update the data dictionary
data = {
    '기초_세럼': multiple_serum,
    '기초_스킨': np.random.choice(skin, size=num_samples),
    '기초_로션': np.random.choice(lotion, size=num_samples),
    '기초_크림': np.random.choice(cream, size=num_samples)
}

# Create a new DataFrame with updated 'Fruit' column
customer_item_sample = pd.DataFrame(data)

# Display the first few rows of the new DataFrame
customer_item_sample.head()
```
