```python
import pandas as pd
```

```python
data = pd.read_csv('./data/젤다의 전설_발더스 게이트_호그와트 레거시_네이버_월간검색트렌드.csv')
data
```

```python
# !pip install matplotlib
# !pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family']='paybooc'
fig= plt.figure(figsize=(15,5))
sns.barplot(x='period', y='searchVolume', hue='keyword',data=data)
```

```python
data.info()
```

```python
# data['period'][0].split("-")[0]
data['year'] = data['period'].apply(lambda x : x.split("-")[0])
data['month'] = data['period'].apply(lambda x : x.split("-")[1])
data
```

```python
data_2023 = data[data['year']=='2023']
data_2023 = data_2023.set_index('period')
data_2023
```

```python
hogwart = data_2023[data_2023['keyword']=='호그와트 레거시']
zelda = data_2023[data_2023['keyword']=='젤다의 전설']
baldurs = data_2023[data_2023['keyword']=='발더스 게이트']
hogwart
```

```python
def growth(dataframe):
    for i in range(len(dataframe)-1):
        try:
            dataframe['growth'][i+1] = (dataframe['searchVolume'][i+1] / dataframe['searchVolume'][i]) * 100
        except:
            dataframe['growth'] = 'not applicable'
    dataframe['growth'] = pd.to_numeric(dataframe['growth'], errors='coerce')
    return dataframe
```

```python
hogwart = growth(hogwart)
zelda = growth(zelda)
baldurs = growth(baldurs)
```
