```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait

service = Service()
options = webdriver.ChromeOptions()
```

- 사당 하나 섹션 클래스 : hlEDJZq7 또는_4J1PtIp8
- 소제목들 클래스 : iO2t5tYt
- 지역 이름 id s-5.1.1 또는 s-5.1

```python
driver = webdriver.Chrome(service=service,options=options)
time.sleep(0.2)


url = "https://namu.wiki/w/%ED%8C%8C%EB%A7%88%EC%9D%98%20%EC%82%AC%EB%8B%B9"
driver.get(url)

titles = driver.find_elements(By.CLASS_NAME,"iO2t5tYt")
shrines = driver.find_elements(By.CLASS_NAME,"_4J1PtIp8")
```

```python
regions_names = []
for title in titles:
    title_text = title.text.strip().replace("\n[편집]","")
    if title_text !='[편집]' and '5.' in title_text and '기타' not in title_text:
        regions_names.append(title_text)
regions_names
```

```python
shrine_explanations = []
shrine_kor_names = []
shrine_eng_names = []
try: 
    for idx,shrine in enumerate(shrines):
        if idx >= 2 and idx < 154:
            shrine_text = shrine.text
            shrine_explanations.append(shrine_text.split("\n")[-1].strip())
            shrine_kor_names.append(shrine_text.split("Lightroot")[0].split("-")[0].strip())
            shrine_eng_names.append(" ".join(shrine_text.split("Lightroot")[0].strip().split("/ ")[1].split(" - ")[0].split()))
            # if "Lightroot" in shrine_text: shrine_explanations.append(shrine_text.split("Lightroot")[-1].strip())
            # else: shrine_explanations.append(shrine_text.split("\n")[-1].strip())
            
except IndexError: pass
```

```python
shrine_names = {key:value for key,value in zip(shrine_kor_names,shrine_eng_names)}
shrine_names
```
