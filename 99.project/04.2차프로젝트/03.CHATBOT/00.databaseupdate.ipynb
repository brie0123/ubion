# 각 사당별 설명 쪼개기

```python
shrine_walkthrough = zelda_df[['shrine_eng','shrine_kor','walkthrough_eng','walkthrough_kor']]
shrine_walkthrough
```

```python
shrine_walkthrough
```

## shrine walkthrough_eng preprocessing

```python
shrine_walkthrough['walkthrough_clear_eng'] = ""    # how to clear
shrine_walkthrough['type'] = ""                 # quest given from the shrine
shrine_walkthrough['location_eng'] = ""             # oral explanation of location

for idx,walk in enumerate(shrine_walkthrough['walkthrough_eng']):
    texts = walk.split(".")
    shrine_walkthrough.loc[idx,'type'] = texts[0].split(",")[1].replace("also known as ","").strip()
    shrine_walkthrough.loc[idx,'location_eng'] = texts[1].strip()
    shrine_walkthrough.loc[idx,'walkthrough_clear_eng'] = ".".join(texts[2:])

shrine_walkthrough
```

## shrine_walkthrough_kor preprocessing

```python
shrine_walkthrough['walkthrough_clear_kor'] = ""    # how to clear_kor
shrine_walkthrough['location_kor'] = ""             # oral explanation of location_kor

for idx,walk in enumerate(shrine_walkthrough['walkthrough_kor']):
    if walk != "이 칸에는 사당과 뿌리에 대한 설명이 들어갑니다.":
        texts = walk.split(".")
        shrine_walkthrough.loc[idx,'location_kor'] = texts[0].strip()
        shrine_walkthrough.loc[idx,'walkthrough_clear_kor'] = ".".join(texts[1:])

shrine_walkthrough
```

## shrine_walkthrough 칼럼 filtering

```python
shrine_walkthrough_filtered = shrine_walkthrough[['shrine_eng','shrine_kor','walkthrough_clear_eng','walkthrough_clear_kor','type']]
shrine_walkthrough_filtered
```

```python
# shrine__walkthrough_filtered db 저장
conn = pymysql.connect(host='localhost', user='root', password='apple123!!', db='zelda', charset='utf8mb4')

try:
    with conn.cursor() as curs:
        sql = """
        CREATE TABLE IF NOT EXISTS shrine_walkthrough (
            shrine_eng VARCHAR(512) PRIMARY KEY,
            shrine_kor VARCHAR(512),
            walkthrough_clear_eng TEXT,
            walkthrough_clear_kor TEXT,
            type VARCHAR(512)
        ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4
        """
        curs.execute(sql)
        conn.commit()
finally:
    conn.close()
```

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:apple123!!@localhost/zelda?charset=utf8mb4')
shrine_walkthrough_filtered.to_sql("shrine_walkthrough", engine, index=False, if_exists="replace")
```

# shrine_location filtering

```python
shrine_location = zelda_df[['shrine_eng','shrine_kor','coordinate','region_eng','region_kor']]
shrine_location['location_eng'] = shrine_walkthrough['location_eng']
shrine_location['location_kor'] = shrine_walkthrough['location_kor']
shrine_location
```

```python
#shrine_location DB 저장
conn = pymysql.connect(host='localhost', user='root', password='apple123!!', db='zelda', charset='utf8mb4')

try:
    with conn.cursor() as curs:
        sql = """
        CREATE TABLE IF NOT EXISTS shrine_location (
            shrine_eng VARCHAR(512) PRIMARY KEY,
            shrine_kor VARCHAR(512),
            coordinate VARCHAR(512),
            region_eng VARCHAR(512),
            region_kor VARCHAR(512),
            location_eng VARCHAR(512),
            location_kor VARCHAR(512)
        ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4
        """
        curs.execute(sql)
        conn.commit()
finally:
    conn.close()
```

```python
engine = create_engine('mysql+pymysql://root:apple123!!@localhost/zelda?charset=utf8mb4')
shrine_location.to_sql("shrine_location", engine, index=False, if_exists="replace")
```
