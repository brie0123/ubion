```python
# 같은 날 평균 기온으로 채우기

import pandas as pd

# Assuming df is a DataFrame with columns 'mmdd' and '최고기온'

key_df = df.groupby('mmdd', as_index=False)['최고기온'].mean()

nulls = df[df['최고기온'].isna()]

for index in nulls.index:
    key = df.loc[index, 'mmdd']
    mean_value = key_df[key_df['mmdd'] == key]['최고기온'].values[0]
    df.loc[index, '최고기온'] = mean_value
```

```python
df.info()
```

```python
df.fillna(0, inplace=True)
df.info()
```

```python
df.rename(columns={'일시':'ds','평균기온':'y'},inplace=True)
df.info()
```

# 머신러닝

```python
df.set_index('ds',inplace=True)
df.index.freq = 'D'
```

```python
from statsmodels.tsa.arima.model import ARIMA

# ARIMA 모델 훈련
model = ARIMA(df['y'], order=(0, 1, 10))  # p, d, q 값은 조정 필요
model_fit = model.fit()
```

```python
submission_df = pd.read_csv('./data/sample_submission.csv')
submission_df['일시'] = pd.to_datetime(submission_df['일시'])

# 예측할 기간 설정
start_date = submission_df['일시'].min()
end_date = submission_df['일시'].max()

# ARIMA 모델을 사용하여 예측
forecast = model_fit.predict(start=start_date, end=end_date, typ='levels')
```

```python
# 예측 결과 추출 및 확인
submission_df['평균기온'] = forecast.values
submission_df2 = submission_df
display(submission_df2.head())

# 예측 결과 저장
submission_df2.to_csv('./baseline_submit2.csv', index=False)
```
